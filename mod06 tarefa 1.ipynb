```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sinasc_df = pd.read_csv('sinasc_RO_2019.csv')

sinasc_df
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.rc('figure', figsize=(10, 5))
sns.barplot(x='CONSULTAS', y='APGAR5', data=sinasc_df, errorbar = None)
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
fig, axes = plt.subplots(2,2, figsize=(10,10))

sns.countplot(ax = axes[0,0], x='SEXO', data=sinasc_df)
sns.countplot(ax = axes[0,1], x='GESTACAO', data=sinasc_df)
sns.countplot(ax = axes[1,0], x='ESTCIVMAE', data=sinasc_df)
sns.countplot(ax = axes[1,1], x='CONSULTAS', data=sinasc_df)
```
